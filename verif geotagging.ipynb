```python
from exif import Image
import os
```

```python
folder = os.getcwd()
image_folder = os.path.join(folder,'photos')
liste_images = os.listdir(image_folder)
len(liste_images)
```

```python
for e in liste_images:
    with open(os.path.join(image_folder, e), 'rb') as image_file:
        image_bytes = image_file.read()
        photo = Image(image_bytes)
        liste_exif = photo.list_all()
        if 'gps_latitude' not in liste_exif:
            print(e)
```
